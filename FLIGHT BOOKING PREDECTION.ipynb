```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['airline'] = le.fit_transform(df['airline'])
df['source_city'] = le.fit_transform(df['source_city'])
df['departure_time'] = le.fit_transform(df['departure_time'])
df['stops'] = le.fit_transform(df['stops'])
df['arrival_time'] = le.fit_transform(df['arrival_time'])
df['destination_city'] = le.fit_transform(df['destination_city'])
df['class'] = le.fit_transform(df['class'])
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
df.drop(['Unnamed: 0'] ,axis = 1, inplace = True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
col_list = []
for col in df.columns:
    if (df[col].dtype != 'object' and  col != "price"):
        col_list.append(col)
        
        
vif = pd.DataFrame()
```

```python
x = df[col_list]
vif['features'] = x.columns
vif['vif'] = [variance_inflation_factor(x.values , i) for i in range(len(x.columns))]
```

```python
print(vif)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(columns = ['price' , 'flight'])
```

```python
X
```

```python
Y = df['price']
```

```python
Y
```

```python
x_train , x_test , y_train , y_test  = train_test_split(X , Y, test_size=0.2 , shuffle=True)
```

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
lr  = LinearRegression()

lr.fit(x_train , y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python

diff = pd.DataFrame(np.c_[y_test ,y_pred] , columns=['actual value' , 'predicted value'])
```

```python
diff
```

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
print(r2_score(y_test , y_pred)*100)

print(mean_absolute_error(y_test , y_pred))

print(np.sqrt(mean_squared_error(y_test , y_pred)))
```

```python
sns.histplot(y_test , label = 'actual')
sns.histplot(y_pred , label = 'predicted')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()

dt.fit(x_train , y_train)

dt_y_pred = dt.predict(x_test)

```

```python
print(r2_score(y_test , dt_y_pred)*100)

print(mean_absolute_error(y_test , dt_y_pred))

print(np.sqrt(mean_squared_error(y_test , dt_y_pred)))
```

```python
sns.distplot(y_test , label = 'actual')
sns.distplot(dt_y_pred , label = 'predicted')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rt = RandomForestRegressor()
```

```python
rt.fit(x_train , y_train)
```

```python
rt_y_pred = rt.predict(x_test)
```

```python
sns.distplot(y_test , label = 'actual')
sns.distplot(rt_y_pred , label = 'predicted')
plt.legend()
plt.show()
```

```python
print("LINEAR REGRESSION")

print(r2_score(y_test , y_pred)*100)

print(mean_absolute_error(y_test , y_pred))

print(np.sqrt(mean_squared_error(y_test , y_pred)) , '\n')


print("DECISION TREE")

print(r2_score(y_test , dt_y_pred)*100)

print(mean_absolute_error(y_test , dt_y_pred))

print(np.sqrt(mean_squared_error(y_test , dt_y_pred)) , '\n')


print("RANDOM FOREST")
print(r2_score(y_test , rt_y_pred)*100)

print(mean_absolute_error(y_test , rt_y_pred))

print(np.sqrt(mean_squared_error(y_test , rt_y_pred)))
```
